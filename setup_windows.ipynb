# Setup the lab for Windows

In the following computational experiment, a Linux environment in Windows will be used.

Please follow the instructions to set up a Ubuntu shell in windows:

- Search: "Turn Windows features on or off"
- Click on "Windows Subsystem for Linux"
- Go to Windows Store and seach "Linux"
- Click on Ubuntu and Launch it. Wait until the program is installed.
- You will have to setup a personal Username and password.

Additional information can be found in this link:

https://medium.com/@sayanghosh_49221/jupyter-notebook-in-windows-subsystem-for-linux-wsl-f075f7ec8691

### Graphical User interface (GUI)
 
In order to visualize the molecules, install Xming in Windows using the link below:
 
https://sourceforge.net/projects/xming/


Once Ubuntu and Xming is installed, run Xming (it will open a small hide icon in the bottom right side of your screen) and run the Ubuntu app.

The Ubuntu app will open as a black screen in which you can only type commands.

### Anaconda for Linux

You have to get Anaconda throw the command line by typing: 
 
> wget http://repo.continuum.io/archive/Anaconda3-2020.02-Linux-x86_64.sh

This will download the installation package of Anaconda to your Linux environment. To install it, type:

> bash Anaconda3-2020.02-Linux-x86_64.sh

Press Enter until it ask you to type yes. Type yes, press Enter and wait until the installation is finished. Once it finished, press Enter one more time and type:

> source .bashrc


### Installation of python packages

For the lab, two python packages (ASE and Pyscf) are required. Type:
For ASE:

> conda install -c conda-forge ase

Type Y, and the package should succesfully installed.

Repeat the process for pyscf by typing:

> pip install -U pyscf

###  Installation of pyberny

PySCF computes the total energy of a molecular and the forces on the atoms In order to use these to optimise the geometry we will use another code availabe from ASE : PyBerny


(<b>NB</b>: this implements a method developed by the famous Quantum Chemist Professor Bernhard Schlegel)


> pip install -U pyberny


### Running the lab

Once everything is setup, you have to download the jupyter notebooks to your Desktop.

To run the jupyter notebooks through the Ubuntu app you first have to move to your Desktop by typing:

> cd /mnt/c/Users/[your Windows username]/Downloads

Once you are in your Desktop, type:

> export DISPLAY=:0

and

> jupyter notebook
 
It will return a link that you have to copy (ctrl+c) and paste it in a webpage browser (e. g. http://localhost:8888/?token=...).

(<b>NB</b>: Do not forget to have Xming running in the background)
